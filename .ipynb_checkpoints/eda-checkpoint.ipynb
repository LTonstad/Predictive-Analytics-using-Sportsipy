# Correcting the **positions** field

## To update non-numeric column to an *int64* datatype

```python
for idx, pos in enumerate(df['field_goals_attempted']):
    if pos is None:
        print(idx, '    ', pos, '    ', type(pos), '    ', df['field_goals_attempted'][idx])
        df['field_goals_attempted'][idx] = np.nan
        
df['field_goals_attempted'] = df['field_goals_attempted'].fillna(0)
df['field_goals_attempted'] = df['field_goals_attempted'].astype('int64')
```

## Checking stats to identify if non-labeled player should be kicker/punter

for idx, pos in enumerate(df['field_goals_attempted']):

## Below is looking into what to do with players that have positions listed for some seasons but an empty string for others (This appears to be because players that did not start at their position for a particular season are not assigned a position on the team)

````python
player_ids = df['player_id'].unique()
for idx, player in enumerate(player_ids):
    player_df = df[df['player_id'] == player]
    positions = player_df['position'].unique()
    print(player_df.name.unique(), '     ', ' '.join(positions).split())
    #df[df['player_id'] == player]['position'] = ' '.join(positions).split()
````
